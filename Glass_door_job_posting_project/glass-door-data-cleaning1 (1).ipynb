```python
import pandas as pd
import re
```

```python
#Import data
df = pd.read_csv('/kaggle/input/data-science-job-posting-on-glassdoor/Uncleaned_DS_jobs.csv')
df
```

```python
#Changing the unit of value of Salary Estate
df['Salary Estimate'] = df['Salary Estimate'].str.replace('K', '000')
df['Salary Estimate']
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x : re.sub(r'[^0-9-]', '', x))
df['Salary Estimate']
```

```python
#removing unnecessary information from values in the Company Name, Job Description column
df['Job Description'] = df['Job Description'].str.replace('\n', ' ', regex = True)
df['Company Name'] = df['Company Name'].str.split('\n').str[0]
df
```

```python
#Check informations
df.info()
```

```python
#Removing unnecessary columns
df = df.drop(['Headquarters', 'Sector', 'Competitors'], axis = 1)
df
```

```python
#Find another value and replace it with unknown
df['Size'].unique()
```

```python
df['Size'] = df['Size'].str.replace('-1', 'Unknown')
```

```python
df['Founded'].unique()
```

```python
#Find another value and replace it with 0
df['Founded'] = df['Founded'].replace(-1, 0)
```

```python
#Find another value and replace it with unknown
df['Type of ownership'].unique()
```

```python
df['Type of ownership'] = df['Type of ownership'].str.replace('-1', 'Unknown')
```

```python
#Find another value and replace it with unknown
df['Industry'] = df['Industry'].str.replace('-1', 'Unknown')
```

```python
df['Revenue'].value_counts()
```

```python
#Find another value and replace it with Non-Applicable 
df['Revenue'] = df['Revenue'].str.replace('-1', 'Unknown / Non-Applicable')
```

```python
#check everything is fine
df
```
